```python
import pandas as pd  # Para manipular dados em tabelas
from google.colab import userdata  # Para upload de arquivos
```

```python
import ipywidgets as widgets
from IPython.display import display

# Lista para armazenar os dados das ruas
ruas = []

# Widgets para entrada de dados
bairro = widgets.Text(placeholder='Bairro')
endereco = widgets.Text(placeholder='Endereço')
numero = widgets.IntText(placeholder='Número')
acessibilidade = widgets.Dropdown(
    options=['Sim', 'Não'],
    value='Sim',
    description='Acessibilidade:'
)

# Botão para adicionar calçada
botao_submit = widgets.Button(description='Adicionar Calçada')

# Widget para exibir o histórico de ruas
historico_output = widgets.Output()

def adicionar_calcada(botao):
    bairro_valor = bairro.value
    endereco_valor = endereco.value
    numero_valor = numero.value
    acessibilidade_valor = acessibilidade.value

    # Adiciona os dados da rua à lista
    ruas.append({
        'bairro': bairro_valor,
        'endereco': endereco_valor,
        'numero': numero_valor,
        'acessibilidade': acessibilidade_valor
    })

    # Atualiza o histórico de ruas no widget de output
    with historico_output:
        historico_output.clear_output()
        print("Histórico de Ruas:")
        for i, rua in enumerate(ruas):
            print(f"{i+1}. Bairro: {rua['bairro']}, Endereço: {rua['endereco']}, Número: {rua['numero']}, Acessibilidade: {rua['acessibilidade']}")

    # Limpa os campos de entrada
    bairro.value = ''
    endereco.value = ''
    numero.value = 0

# Associa a função ao botão
botao_submit.on_click(adicionar_calcada)

# Exibe os widgets
display(widgets.VBox([bairro, endereco, numero, acessibilidade, botao_submit, historico_output]))
```
